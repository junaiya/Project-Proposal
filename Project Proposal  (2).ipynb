# To what extent do different forms of social activity connections correlate with health and subjective well-being outcomes among Canadians?

__Focus:__ The correlation between types of social activity connections and two main outcomes: health (physical and/or mental) and subjective well-being (happiness or life satisfaction).

__Parameter of Interest:__ The strength and direction of the correlation between social connections and well-being measures in the Canadian population.

__Variables:__
- Independent Variables (Frequency of different types of Social Activity Connections):
- Board Games
- Text/Messaging Apps 
- Phone Call
- Video Call 
- Face-to-face Conversations
- Religious Service 
- Coffee Date
- Volunteering
- Computer Games
- Discussion Group  
- Group Exercise 
- Video Chat
- Chores 
- Hug
- Kiss
- Letter/Email
- Work Meeting 
- New Friend
- Sex
- Receiving a “check in” message 
- Visit Family/Friends
- Walk 


__Dependent Variables (Health and Well-being Outcomes):__
- Self-reported health status 
- Mental health score
- Subjective happiness rating


__Exploration Method:__

Plots and Visualizations:
- Bar charts for the average well-being scores across different levels of social connection types to visualize general trends.
- Scatter plots for each type of social connection against well-being scores to show individual-level correlations.

__Summary Statistics:__ Means, medians, and standard deviations for well-being across groups with different levels of social connections. Correlation coefficients to measure the strength of relationships between connection types and outcomes.

__Justification:__

Bar Charts: 
- Good for summarizing categorical variables like connection type and comparing average outcomes.

Scatter Plots: 
- Useful for identifying the direction and strength of correlations between continuous variables (e.g., frequency of social interactions and subjective well-being).

__Analyses:__

Correlation:
A correlation analysis to examine the association between different forms of social activity connections and health and subjective well-being outcomes. This can be done by calculating correlation coefficients and creating scatter plots to visualize the relationship between variables. 

Simple Linear Regression:
A simple linear regression to predict health and subjective well-being outcomes based on different forms of social activity connections. This involves fitting a linear equation to the data, where the form of social connection is the predictor variable, and the outcome (health/well-being) is the response variable. 

Two-Sample Group Comparisons: 
Comparing health and subjective well-being outcomes between groups with different levels or types of social connections. By dividing individuals into groups based on their social connection strength, type, or frequency and then examining differences in their health and well-being scores, employing bootstrapping to assess if the observed differences between groups are statistically significant and utilizing indicator variables in regression models to represent these different groups.


__Hypothesis:__ Higher levels of social connections are positively correlated with improved health and subjective well-being outcomes. 

__Relevance to Research Question:__ Confirming this hypothesis would indicate that implementing different types of social connections is associated with better health and happiness outcomes. This insight would support campaigns highlighting the health benefits of strong social networks, advocating for policies that promote community engagement and social support structures.
