# To what extent do different forms of social activity connections correlate with health and subjective well-being outcomes among Canadians?

__Focus:__ The correlation between types of social activity connections and two main outcomes: health (physical and/or mental) and subjective well-being (happiness or life satisfaction).

__Parameter of Interest:__ The strength and direction of the correlation between social connections and well-being measures in the Canadian population.

__Variables:__

__Independent Variables (Frequency of different types of Social Activity Connections):__

- Board Games
- Text/Messaging Apps 
- Phone Call
- Video Call 
- Face-to-face Conversations
- Religious Service 
- Coffee Date
- Volunteering
- Computer Games
- Discussion Group  
- Group Exercise 
- Video Chat
- Chores 
- Hug
- Kiss
- Letter/Email
- Work Meeting 
- New Friend
- Sex
- Receiving a “check in” message 
- Visit Family/Friends
- Walk 


__Dependent Variables (Health and Well-being Outcomes):__
- Self-reported health status 
- Mental health score
- Subjective happiness rating


__Exploration Method:__

Plots and Visualizations:
- Bar charts for the average well-being scores across different levels of social connection types to visualize general trends.
- Scatter plots for each type of social connection against well-being scores to show individual-level correlations.

__Summary Statistics:__ Means, medians, and standard deviations for well-being across groups with different levels of social connections. Correlation coefficients to measure the strength of relationships between connection types and outcomes.

__Justification:__

Bar Charts: 
- Good for summarizing categorical variables like connection type and comparing average outcomes.

Scatter Plots: 
- Useful for identifying the direction and strength of correlations between continuous variables (e.g., frequency of social interactions and subjective well-being).

__Analyses:__

Correlation:
A correlation analysis to examine the association between different forms of social activity connections and health and subjective well-being outcomes. This can be done by calculating correlation coefficients and creating scatter plots to visualize the relationship between variables. 

Simple Linear Regression:
A simple linear regression to predict health and subjective well-being outcomes based on different forms of social activity connections. This involves fitting a linear equation to the data, where the form of social connection is the predictor variable, and the outcome (health/well-being) is the response variable. 

Two-Sample Group Comparisons: 
Comparing health and subjective well-being outcomes between groups with different levels or types of social connections. By dividing individuals into groups based on their social connection strength, type, or frequency and then examining differences in their health and well-being scores, employing bootstrapping to assess if the observed differences between groups are statistically significant and utilizing indicator variables in regression models to represent these different groups.


__Hypothesis:__ Higher levels of social connections are positively correlated with improved health and subjective well-being outcomes. 

__Relevance to Research Question:__ Confirming this hypothesis would indicate that implementing different types of social connections is associated with better health and happiness outcomes.

**Research Question:** To what extent do different forms of COVID-19 prevention behaviors correlate with health and subjective well-being outcomes among Canadians?

**Focus:** The correlation between types of COVID-19 prevention behaviors and two main outcomes: health and subjective well-being.

**Parameter of Interest:** The strength and direction of the correlation between COVID-19 prevention practices and well-being measures in the Canadian population.

**Variables:**

**Independent Variables (Frequency of COVID-19 Prevention Behaviors):**
- Avoiding non-essential trips in the community
- Physical distancing (maintaining 2 metres from others)
- Hand-washing frequency
- Limiting indoor socializing to immediate household members
- Mask-wearing in public
- Reducing interactions with others
- COVID-19 vaccination status

**Dependent Variables (Health and Well-being Outcomes):**
- Self-reported health status
- Mental health score
- Subjective happiness rating

**Exploration Method:**

1. **Plots and Visualizations:**
   - **Bar charts** for average well-being scores across different levels of COVID-19 prevention behaviors to visualize general trends.
   - **Scatter plots** for each prevention behavior against well-being scores to show individual-level correlations.

2. **Summary Statistics:**
   - Means, medians, and standard deviations for well-being scores across groups with different levels of COVID-19 prevention behaviors.
   - Correlation coefficients to measure the strength of relationships between behaviors and outcomes.

**Justification:**

- **Bar Charts:** Ideal for summarizing categorical behaviors (e.g., vaccination status, mask-wearing) and comparing average outcomes.
- **Scatter Plots:** Useful for identifying the direction and strength of correlations between continuous variables (e.g., frequency of behaviors and subjective well-being).

**Analyses:**

1. **Correlation Analysis:** Examining associations between COVID-19 prevention behaviors and health and subjective well-being outcomes through correlation coefficients and scatter plots.

2. **Simple Linear Regression:** Predicting health and subjective well-being outcomes based on various COVID-19 prevention behaviors, where each prevention behavior is the predictor variable and health/well-being is the response variable.

3. **Two-Sample Group Comparisons:** Comparing health and well-being outcomes between groups with varying levels of prevention adherence. For example, comparing outcomes between vaccinated and non-vaccinated individuals and those who always wear masks versus those who do not.

**Hypothesis:** Higher adherence to COVID-19 prevention behaviors is positively correlated with improved health and subjective well-being outcomes.

**Relevance to Research Question:** Confirming this hypothesis would suggest that engaging in COVID-19 prevention behaviors is associated with better health and happiness outcomes.

**Research Question:** To what extent do different forms and intensities of loneliness correlate with health and subjective well-being outcomes among Canadians?

**Focus:** The correlation between various loneliness measures and two main outcomes: health and subjective well-being.

**Parameter of Interest:** The strength and direction of the correlation between different dimensions of loneliness and well-being measures in the Canadian population.

**Variables:**

**Independent Variables (Loneliness Measures):**
- Changes in loneliness intensity compared to pre-COVID-19.
- Feeling close to enough people.
- General sense of emptiness.
- Missing having people around.
- Feeling often rejected.
- Having people to rely on during problems.
- Having people to trust completely.
- Direct feelings of loneliness in the past week.
- Perception of others’ awareness of one’s loneliness.
- Frequency of feeling a lack of companionship.
- Frequency of feeling isolated.
- Frequency of feeling left out.

**Dependent Variables (Health and Well-being Outcomes):**
- Self-reported health status
- Mental health score
- Subjective happiness rating

**Exploration Method:**

1. **Plots and Visualizations:**
   - **Bar charts** for average well-being scores across different levels of loneliness indicators to visualize general trends.
   - **Scatter plots** for each loneliness measure against well-being scores to show individual-level correlations.

2. **Summary Statistics:**
   - Means, medians, and standard deviations for well-being scores across groups with different loneliness intensities.
   - Correlation coefficients to measure the strength of relationships between loneliness measures and outcomes.

**Justification:**

- **Bar Charts:** Ideal for summarizing categorical variables like loneliness intensity and comparing average health and happiness outcomes.
- **Scatter Plots:** Useful for identifying the direction and strength of correlations between continuous variables (e.g., frequency of loneliness and subjective well-being).

**Analyses:**

1. **Correlation Analysis:** Assessing associations between loneliness dimensions and health/well-being through correlation coefficients and scatter plots.

2. **Simple Linear Regression:** Predicting health and subjective well-being outcomes based on various loneliness measures, with each loneliness indicator serving as a predictor and health/well-being as the response variable.

3. **Two-Sample Group Comparisons:** Comparing health and well-being outcomes between groups with high and low loneliness indicators (e.g., those with vs. without a sense of companionship) and examining differences in health and well-being scores.

**Hypothesis:** Higher levels of loneliness are negatively correlated with health and subjective well-being outcomes.

**Relevance to Research Question:** Confirming this hypothesis would highlight the impact of loneliness on Canadians' health and happiness, suggesting that addressing loneliness could improve well-being outcomes and inform supportive policies.
